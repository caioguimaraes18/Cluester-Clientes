```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime


df = pd.read_csv('../Dataset/Segmentacao.csv')

df.head()
```

```python
def plot_segmento(x,y,data):
    
    sns.set(palette='muted', color_codes=True, style="whitegrid")
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=x, y=y, hue="Segmento", data=data, size="Segmento", size_order=["Premium", "Master", "Business", "Inativo"], sizes=(20, 200))
    plt.title("Segmentação de Clientes")
    plt.xlabel(x)
    plt.ylabel(y)
    plt.legend(title="Segmento", loc='upper right')
    plt.grid(True)
    plt.show()
```

```python
plot_segmento("Recencia", "Frequencia", df)
```

```python
plot_segmento('Frequencia', 'Receita', df)  
```

```python
plot_segmento("Recencia", "Receita", df)
```

```python
def plot_perc(st, dados):
    plt.figure(figsize=(10, 6))
    
    g = sns.countplot(x= st, data=dados, orient='horizontal', palette='Blues_r')
    g.set_ylabel('Contagem', fontsize=14)
    
    sizes = []
    
    for i in g.patches:
        h = i.get_height()
        sizes.append(h)
        g.text(i.get_x() + i.get_width()/1.6,
               h + 200,
               '{:1.2f}%'.format((h/len(dados))*100),
               ha = 'center', va='bottom',
               fontsize=12)
        
        g.set_ylim(0, max(sizes)*1.1)        
```

```python
plot_perc('Segmento', df)
```

#### Distribuição dos Segmentos de Clientes


O gráfico acima apresenta a distribuição percentual de clientes por segmento, com base na análise RFM (Recência, Frequência e Monetário). É possível observar:

- Inativos somam 12,96%, um grupo que não realiza compras há bastante tempo e pode ser alvo de campanhas de reativação.

- Business representa a maior parte da base, com 51,92% dos clientes. São clientes com bom nível de atividade, frequência ou valor de compra.

- Master vem em seguida com 34,52%, indicando clientes ainda mais engajados, porém em menor número.

- Premium, apesar de representar apenas 0,61% da base, são clientes extremamente valiosos e fiéis, com alta frequência e receita — o grupo mais estratégico para retenção.

#### Conclusão com Estratégias por Segmento

Cada segmento exige uma abordagem específica:

- Premium: foco em retenção com benefícios exclusivos, atendimento personalizado e ofertas VIP.

- Master: foco em fidelização, com recomendações personalizadas e campanhas de upsell.

- Business: foco em engajamento, usando promoções e incentivos à recompra.

- Inativo: foco em reativação, com cupons de retorno, comunicações diretas e ofertas pontuais.

Com essa segmentação, é possível direcionar ações mais eficientes e personalizadas, aumentando a performance e o relacionamento com os clientes.
